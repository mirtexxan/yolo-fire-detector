## 5. Verifica Dataset

Controlliamo che il dataset sia stato creato correttamente.

````python
import os
dataset_path = "dataset"

if os.path.exists(dataset_path):
    print("✅ Dataset creato:")
    for root, dirs, files in os.walk(dataset_path):
        level = root.replace(dataset_path, '').count(os.sep)
        indent = ' ' * 2 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 2 * (level + 1)
        for file in files[:5]:  # mostra solo primi 5 file
            print(f"{subindent}{file}")
        if len(files) > 5:
            print(f"{subindent}... e altri {len(files)-5} file")
else:
    print("❌ Dataset non trovato")
````

## 7. Verifica Risultati Training

Controlliamo i risultati del training.

````python
import os
results_path = "runs/detect/fire_detector_runs/train"

if os.path.exists(results_path):
    print("✅ Training completato!")
    print("\nFile generati:")
    for file in os.listdir(results_path):
        if os.path.isfile(os.path.join(results_path, file)):
            print(f"  - {file}")
    
    weights_path = os.path.join(results_path, "weights")
    if os.path.exists(weights_path):
        print(f"\nModelli in {weights_path}:")
        for file in os.listdir(weights_path):
            print(f"  - {file}")
else:
    print("❌ Training non completato o risultati non trovati")
````

## 8. Scarica Modello Finale

Scarica il modello addestrato e le impostazioni per usarlo su un'altra macchina.

````python
from google.colab import files
import os

final_export = "runs/detect/fire_detector_runs/train/final_export"

if os.path.exists(final_export):
    print("📁 File da scaricare:")
    
    for file in os.listdir(final_export):
        file_path = os.path.join(final_export, file)
        print(f"  - {file}")
        files.download(file_path)
    
    print("\n✅ Download completato!")
    print("Ora puoi usare questi file su qualsiasi macchina per detection.")
else:
    print("❌ Cartella final_export non trovata")
    print("Verifica che il training sia completato correttamente.")
````
